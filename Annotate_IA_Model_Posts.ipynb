```python
import pandas as pd
import re
import emoji
```

```python
# Charger les fichiers Excel
posts_df = pd.read_excel('/content/Posts.xlsx')

# Afficher DataFrame "Comments"
posts_df
```

```python
def clean_text(text):
    text = emoji.replace_emoji(text, replace=" ")  # Supprimer les emojis
    text = re.sub(r'http\S+ | htps\S+', " ", str(text))  # Supprimer les hyperliens
    text = re.sub(r'((www\.[^\s]+)|(https?://[^\s]+))', " ", str(text))  # Supprimer les URL
    text = re.sub(r'@\S+', '', str(text))  # Supprimer les mots commençant par @
    text = text.replace("_", " ").replace("#", "")  # Supprimer # et _
    text = text.replace("'", " ")  # Supprimer '
    text = re.sub(r'\. | , | ، | ؛', " ", text)  # Supprimer les ponctuations

    # Supprimer les voyelles courtes arabes (حركات)
    harakat = "[\u064B-\u0652]"  # Comprend
    text = re.sub(harakat, '', text)

    text = re.sub(r"(.)\1{2,}", r"\1", text)  # Supprimer les caractères consécutifs en double
    text = text.replace('\n', " ").replace('/', " ")  # Supprimer sauts de ligne et /
    text = re.sub(r'[^\w\s]', " ", text)  # Supprimer les caractères spéciaux

    return text.strip()

posts_df["Contents"] = posts_df["Contents"].apply(clean_text)

# Remplacer les valeurs nulles par une chaîne vide
posts_df["Contents"] = posts_df["Contents"].fillna('')

posts_df = posts_df.dropna(subset=["Contents"])
posts_df = posts_df[posts_df["Contents"].str.strip() != ""]

posts_df
```

```python
from llama_cpp import Llama

llm = Llama(
    model_path="/content/models/DeepSeek-R1-Distill-Qwen-7B-Q6_K.gguf",
    n_gpu_layers=-1,  # Utilise toutes les couches sur le GPU
    use_mlock=False,
    n_ctx=2048,
    verbose=False
)
```

```python
system_prompt = (
    "You are a highly intelligent multilingual assistant specialized in **telecom post classification**. "
    "Your task is to analyze the content of each user post, identified by a **unique ID**, and classify it into one or more **relevant subcategories** based on the issues, topics mentioned.\n\n"

    "### 🎯 Classification Subcategories:\n"
    "- **Culture & Celebration** (e.g., religious greetings, Supplications, blessings, or religious holiday mentions, Yennayer, Aïd, fêtes nationales non religieuses)\n"
    "- **Sport** (e.g., posts mentioning football matches, footbal sport, favorite teams, or sports-related events)\n"
    "- **Product** (e.g., feedback on mobile offers, plans, devices, SIM cards, or new products or offers)\n"
    "- **Promotion** (e.g., promotional campaigns, ...)\n"
    "- **Network** (e.g., issues with mobile coverage, internet speed, dropped calls, or outages)\n"
    "- **Survey** (e.g., responses to satisfaction surveys, polls, or rating requests)\n"
    "- **Event** (e.g., comments about concerts, sponsored festivals, or public gatherings by the brand, actions solidaires de l’opérateur, soutien à une cause)\n"
    "- **Contest & Games** (e.g., participation à un jeu-concours, tirages au sort)\n"
    "- **Other** (e.g., any content that does not fit the above categories: memes, general remarks, off-topic content, etc.)\n\n"

    "### 🧠 Classification Instructions:\n"
    "1. Carefully read the content of each post.\n"
    "2. Identify and assign the most appropriate subcategory or subcategories based on the topic(s).\n"
    "3. You may assign **multiple subcategories** when multiple themes are mentioned.\n"
    "4. Avoid assigning only one category if more are clearly present in the content.\n"

    "### 📤 Output Format:\n"
    "Respond **strictly** using the following JSON structure (no extra explanations):\n\n"
    "{\n"
    "  \"45\": {\n"
    "    \"categories\": [\"Culture & Celebration\", \"Event\"]\n"
    "  },\n"
    "  \"66\": {\n"
    "    \"categories\": [\"Promotion\", \"Product\", \"Network\"]\n"
    "  },\n"
    "  \"69\": {\n"
    "    \"categories\": [\"Network\"]\n"
    "  }\n"
    "}\n\n"

    "⚠️ **Important Notes**:\n"
    "- Do **not** include any commentary or reasoning in your output.\n"
    "- Use only the exact subcategory labels provided above — no improvisation.\n"
    "- Ensure your output is in **valid JSON format**.\n"
    "- Each post ID must be assigned **at least one** subcategory.\n\n"

    "### 🌐 Supported Languages:\n"
    "- Arabic (العربية)\n"
    "- Algerian Darija (الدارجة الجزائرية)\n"
    "- French (Français)\n"
    "- English\n\n"

    "Begin the classification process now."
)
```

```python
# Extraire uniquement les contents dont l'ID est entre 133 et 183 inclus
test_df = posts_df[(posts_df["ID"] >= 133) & (posts_df["ID"] <= 183)].copy()

# Aperçu du DataFrame filtré
test_df
```

```python
import json
import re
from tqdm import tqdm
from time import sleep

# Boucle d’inférence
results = {}

for idx, row in test_df.iterrows():
    content_id = str(row['ID'])
    content_text = str(row['Contents'])

    prompt = (
        system_prompt
        + f"\n\nClassify the following content of the post. Respond using ONLY the JSON format.\n"
        + f"ID: \"{content_id}\"\n"
        + f"Contents: \"{content_text}\"\n"
        + "\nDO NOT limit to one category if the content contains multiple concerns.\n"
        + "\nReturn the result in this exact structure:\n"
        + "{\n"
        + f"  \"{content_id}\": {{\n"
        + "    \"categories\": [\"Category A\", \"Category B\"]\n"
        + "  }\n"
        + "}\n"
        + "\n⚠️ DO NOT COMMENT. DO NOT THINK. ONLY RETURN STRICT JSON."
    )

    try:
        output = llm(prompt, max_tokens=256, temperature=0.0)
        raw_response = output["choices"][0]["text"].strip()

        # Extraction du JSON
        json_match = re.search(r"\{.*\}", raw_response, re.DOTALL)
        if json_match:
            json_str = json_match.group()
            parsed = json.loads(json_str)
            results.update(parsed)
            print(json.dumps(parsed, indent=2, ensure_ascii=False))  # 👈 Affichage formaté JSON uniquement
        else:
            raise ValueError("No JSON found")

    except Exception:
        fallback = {
            content_id: {
                "categories": ["Other"]
            }
        }
        results.update(fallback)
        print(json.dumps(fallback, indent=2, ensure_ascii=False))

    sleep(0.5)
```

```python
# Sauvegarder en JSON
json_path = "/content/Results/classified_posts.json"
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(results, f, indent=2, ensure_ascii=False)
```

```python
import pandas as pd
import json

# Lire le fichier JSON depuis le disque
with open("/content/Results/classified_posts.json", "r", encoding="utf-8") as f:
    classification_dict = json.load(f)

posts_df = pd.read_excel('/content/Posts.xlsx')

# Liste des catégories
category_list = [
    "Culture & Celebration",
    "Sport",
    "Product",
    "Promotion",
    "Network",
    "Survey",
    "Event",
    "Contest & Games",
    "Other"
]


# Initialiser toutes les colonnes de catégories à 0
for cat in category_list:
    if cat not in posts_df.columns:
        posts_df[cat] = 0

# Remplir en fonction des catégories détectées
for content_id_str, data in classification_dict.items():
    content_id = int(content_id_str)
    categories = data.get("categories", [])
    for category in categories:
        if category in category_list:
            posts_df.loc[posts_df["ID"] == content_id, category] = 1

# Sauvegarder le nouveau DataFrame
posts_df.to_csv("/content/Results/posts_df_classified.csv", index=False)

# Afficher un aperçu
posts_df
```
